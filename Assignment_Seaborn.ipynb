```python
# Seaborn is a powerful visualization library in Python built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn, along with their uses:

# Scatter Plot (sns.scatterplot):

# Use: A scatter plot is used to display the relationship between two continuous variables. It helps in identifying correlations, patterns, and potential outliers.
# Example: Visualizing the relationship between the height and weight of individuals.
# Histogram (sns.histplot):

# Use: A histogram is used to visualize the distribution of a single continuous variable. It helps in understanding the frequency distribution, central tendency, and spread of the data.
# Example: Showing the distribution of exam scores for a class of students.
# Box Plot (sns.boxplot):

# Use: A box plot (or whisker plot) is used to visualize the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps in identifying outliers and comparing distributions across different categories.
# Example: Comparing the test scores across different groups of students.
# Heatmap (sns.heatmap):

# Use: A heatmap is used to visualize data in matrix form, where individual values are represented by color. It's particularly useful for displaying the correlation matrix, helping to identify relationships between multiple variables.
# Example: Showing the correlation matrix of various financial indicators.
# Pair Plot (sns.pairplot):

# Use: A pair plot (or scatterplot matrix) is used to visualize the pairwise relationships between different variables in a dataset. It plots scatter plots for each pair of variables and histograms or kernel density plots for the distribution of each variable.
# Example: Analyzing the relationships between different features in the famous Iris dataset.
# These plots provide versatile tools for exploratory data analysis, enabling users to uncover insights and patterns in their data effectively.

```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age by Passenger Class')

# Create the second box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare by Passenger Class')
```

```python
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

```
